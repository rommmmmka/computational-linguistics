```python
START = "<!-- BOOK_BEGIN -->"
STOP = "<!-- BOOK_END -->"

if not os.path.exists("results/texts"):
    os.makedirs("results/texts")
    
for text_i, url in enumerate(urls):
    print(f"{str(text_i + 1).zfill(2)}/{len(urls)}")
    
    full_page = requests.get(url)
    full_page.encoding = full_page.apparent_encoding
    book_html = full_page.text[
        full_page.text.find(START) + len(START) : full_page.text.find(STOP)
    ]
    
    soup = BeautifulSoup(book_html, "html.parser")
    book_text = ""
    for part in soup.find_all("p"):
        if part.text == "\xa0" or part.text.isnumeric():
            continue
        book_text += part.text + "\n"
        
    with open(f"results/texts/{str(text_i + 1).zfill(2)}.txt", mode="w", encoding="utf-8") as file:
        file.write(book_text)
```
